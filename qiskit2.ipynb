```python
from qiskit import *
```

```python
from qiskit.visualization import *
```

```python
n = 8
```

```python
qc1 = QuantumCircuit(n,n)
for i in range(n):
    qc1.measure(i,i)
```

```python
qc1.draw('mpl')
```

```python
qc2 = QuantumCircuit(n)
```

```python
qc2.x(7)
qc2.draw('mpl')
qc = qc1+qc2
qc.draw('mpl')
qcc = qc2+qc1
qcc.draw('mpl')
```

```python
qc.draw('mpl')
```

```python
sim = Aer.get_backend('qasm_simulator')
qobj = assemble(qcc)
counts = sim.run(qobj).result().get_counts()
plot_histogram(counts)
```

```python
m = 6
```

```python
age = QuantumCircuit(m,m)
age.x(1)
age.x(2)
age.x(5)
age.draw('mpl')
for i in range(m):
    age.measure(i,i)
age.draw('mpl')    
qobj = assemble(age)
result = sim.run(qobj).result()
counts = result.get_counts()
plot_histogram(counts)
    
```
