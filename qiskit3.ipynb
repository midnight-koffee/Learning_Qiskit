```python
from qiskit import *
from qiskit.visualization import *
```

```python
qc = QuantumCircuit(2,2)
qc.x(0)
qc.barrier()
qc.cx(0,1)
```

```python
qc.measure(1,1)
```

```python
qc.draw()
```

```python
qc.draw('mpl')
```

```python
qc.measure(0,0)
```

```python
qc.draw('mpl')
```

```python
sim = Aer.get_backend('qasm_simulator')
```

```python
qobj = assemble(qc)
```

```python
result = sim.run(qobj).result()
```

```python
counts = result.get_counts()
```

```python
plot_histogram(counts)
```

```python
qc_ha = QuantumCircuit(4,2)
#qc_ha.x(0)
qc_ha.x(1)
qc_ha.barrier()
qc_ha.cx(0,2)
qc_ha.cx(1,2)
qc_ha.ccx(0,1,3)
qc_ha.barrier()
```

```python
qc_ha.draw('mpl')
```

```python
qc_ha.measure(2,0)
qc_ha.measure(3,1)
```

```python
qc_ha.draw('mpl')
```

```python
qobj = assemble(qc_ha)
```

```python
result = sim.run(qobj).result()
```

```python
counts = result.get_counts()
```

```python
plot_histogram(counts)
```
